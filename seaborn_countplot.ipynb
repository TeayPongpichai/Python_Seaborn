```python
df['class'].value_counts()
```

```python
df['class'].value_counts().index
```

```python
sns.countplot(x='class',data=df,order=df['class'].value_counts().index)
```

```python
#p = ['maroon','lightgray','maroon','deepskyblue','maroon','maroon','maroon']
# p = ['deepskyblue' if v > 40 else 'lightgray' for v in df['class'].value_counts()]
p = ['deepskyblue' if v == 'pickup' else 'lightgray'
    for v in df['class'].value_counts().index]
sns.countplot(x='class',data=df,palette=p,order=df['class'].value_counts().index)
```

```python
avg = df['class'].value_counts().mean()
sns.countplot(x='class',data=df,color='lightgray',order=df['class'].value_counts().index)
plt.axhline(y=avg,linestyle='--',color='black')
```

```python
avg = df['class'].value_counts().mean()
sns.countplot(y='class',data=df,color='lightgray',order=df['class'].value_counts().index)
plt.axvline(x=avg,linestyle='--',color='black')
```

```python
avg = df['class'].value_counts().mean()
colors = ['tomato' if (x > avg) else 'lightgray' for x in df['class'].value_counts()]
colors
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='class',data = df,palette=colors,order=df['class'].value_counts().index)
plt.axhline(y=avg,linestyle='--',color='deepskyblue')
```
